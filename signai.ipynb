```python
import kagglehub

# Download latest version for asl-alphabet
asl_alphabet_path = kagglehub.dataset_download("grassknoted/asl-alphabet")

print("Path to ASL Alphabet dataset files:", asl_alphabet_path)
```

```python
import kagglehub

# Download latest version for sign-language-mnist
sign_language_mnist_path = kagglehub.dataset_download("datamunge/sign-language-mnist")

print("Path to Sign Language MNIST dataset files:", sign_language_mnist_path)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_gen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_data = train_gen.flow_from_directory(
    asl_alphabet_path + "/asl_alphabet_train/asl_alphabet_train",
    target_size=(64,64),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_data = train_gen.flow_from_directory(
    asl_alphabet_path + "/asl_alphabet_train/asl_alphabet_train",
    target_size=(64,64),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(
    Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(64,64,3)
    )
)

model.add(
    MaxPooling2D(2,2)
)

model.add(
    Conv2D(
        64,
        (3,3),
        activation='relu'
    )
)

model.add(
    MaxPooling2D(2,2)
)

model.add(
    Flatten()
)

model.add(
    Dense(
        128,
        activation='relu'
    )
)

model.add(
    Dense(
        29,
        activation='softmax'
    )
)
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
model.save(
    "asl_model.h5"
)
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import pickle
import os
from sklearn.preprocessing import LabelEncoder

# Load model
model = load_model("asl_model.h5")

# Check if label_encoder.pkl exists, if not, create and save it
if not os.path.exists("label_encoder.pkl"):
    # Assuming train_data (from cell eQrRIYlljJvb) is available in kernel state
    # Reconstruct class names from train_data.class_indices
    # train_data.class_indices is a dict: {'class_name': integer_label}
    # We need the ordered list of class names to fit the LabelEncoder
    class_names_in_order = [class_name for class_name, index in sorted(train_data.class_indices.items(), key=lambda item: item[1])]

    encoder = LabelEncoder()
    encoder.fit(class_names_in_order) # Fit the encoder with the ordered class names

    with open("label_encoder.pkl", "wb") as f:
        pickle.dump(encoder, f)
    print("label_encoder.pkl created and saved.")

# Load label encoder
with open("label_encoder.pkl", "rb") as f:
    encoder = pickle.load(f)

# Read image - !!! IMPORTANT: Ensure 'A_test.jpg' exists in your environment or provide the correct path !!!
img = cv2.imread("A_test.jpg")

# Resize
img = cv2.resize(img, (64, 64))

# Normalize
img = img / 255.0

# Add batch dimension
img = np.expand_dims(img, axis=0)

# Predict
pred = model.predict(img)

pred_class = np.argmax(pred)

label = encoder.inverse_transform([pred_class])

confidence = np.max(pred)

print("Predicted Sign:", label[0])
print("Confidence:", confidence)
```
