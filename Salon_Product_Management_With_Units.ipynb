# Salon Product Management System (with Units)
This notebook creates a simple GUI application for managing salon product data using Tkinter. It includes a Unit field for quantity to differentiate hair (ml) and skin (box) products.

```python

import tkinter as tk
from tkinter import messagebox
import pandas as pd
import os

# Define Excel file path
file_path = "Salon_Products_With_Units.xlsx"

# Function to determine the unit based on product name
def determine_unit(product_name):
    if "hair" in product_name.lower():
        return "ml"
    elif "skin" in product_name.lower():
        return "box"
    else:
        return "unit"

# Function to submit data
def submit_data():
    date = date_entry.get()
    product_id = product_id_entry.get()
    product_name = product_name_entry.get()
    quantity = quantity_entry.get()
    price = price_entry.get()
    
    if not (date and product_id and product_name and quantity and price):
        messagebox.showerror("Error", "All fields must be filled!")
        return
    
    try:
        quantity = float(quantity)
        price = float(price)
        total = quantity * price
    except ValueError:
        messagebox.showerror("Error", "Quantity and Price must be numbers!")
        return

    # Determine unit based on product name
    unit = determine_unit(product_name)

    # Create or append data to Excel
    new_data = pd.DataFrame([[date, product_id, product_name, quantity, unit, price, total]],
                            columns=["Date", "Product ID", "Product Name", "Quantity", "Unit", "Price", "Total"])
    if not os.path.exists(file_path):
        new_data.to_excel(file_path, index=False)
    else:
        existing_data = pd.read_excel(file_path)
        combined_data = pd.concat([existing_data, new_data], ignore_index=True)
        combined_data.to_excel(file_path, index=False)

    messagebox.showinfo("Success", "Data submitted successfully!")
    clear_fields()

# Function to clear input fields
def clear_fields():
    date_entry.delete(0, tk.END)
    product_id_entry.delete(0, tk.END)
    product_name_entry.delete(0, tk.END)
    quantity_entry.delete(0, tk.END)
    price_entry.delete(0, tk.END)
    total_label.config(text="0")

# Function to delete Excel file
def delete_data():
    if os.path.exists(file_path):
        os.remove(file_path)
        messagebox.showinfo("Success", "Data deleted successfully!")
    else:
        messagebox.showerror("Error", "No data file exists!")

# Create Tkinter GUI
root = tk.Tk()
root.title("Salon Product Management (With Units)")
root.geometry("600x450")
root.configure(bg="#f2f2f2")

# Labels and entry fields
tk.Label(root, text="Date (DD/MM/YYYY):", bg="#f2f2f2").grid(row=0, column=0, padx=10, pady=5, sticky=tk.W)
date_entry = tk.Entry(root)
date_entry.grid(row=0, column=1, pady=5)

tk.Label(root, text="Product ID:", bg="#f2f2f2").grid(row=1, column=0, padx=10, pady=5, sticky=tk.W)
product_id_entry = tk.Entry(root)
product_id_entry.grid(row=1, column=1, pady=5)

tk.Label(root, text="Product Name:", bg="#f2f2f2").grid(row=2, column=0, padx=10, pady=5, sticky=tk.W)
product_name_entry = tk.Entry(root)
product_name_entry.grid(row=2, column=1, pady=5)

tk.Label(root, text="Quantity:", bg="#f2f2f2").grid(row=3, column=0, padx=10, pady=5, sticky=tk.W)
quantity_entry = tk.Entry(root)
quantity_entry.grid(row=3, column=1, pady=5)

tk.Label(root, text="Price:", bg="#f2f2f2").grid(row=4, column=0, padx=10, pady=5, sticky=tk.W)
price_entry = tk.Entry(root)
price_entry.grid(row=4, column=1, pady=5)

tk.Label(root, text="Total:", bg="#f2f2f2").grid(row=5, column=0, padx=10, pady=5, sticky=tk.W)
total_label = tk.Label(root, text="0", bg="#ffffff", width=20, anchor=tk.W)
total_label.grid(row=5, column=1, pady=5)

# Buttons
submit_button = tk.Button(root, text="Submit", command=submit_data, bg="#4CAF50", fg="white", width=12)
submit_button.grid(row=6, column=0, pady=10, padx=10)

clear_button = tk.Button(root, text="Clear", command=clear_fields, bg="#FFC107", fg="black", width=12)
clear_button.grid(row=6, column=1, pady=10)

delete_button = tk.Button(root, text="Delete Data", command=delete_data, bg="#F44336", fg="white", width=12)
delete_button.grid(row=7, column=0, pady=10, padx=10)

exit_button = tk.Button(root, text="Exit", command=root.quit, bg="#607D8B", fg="white", width=12)
exit_button.grid(row=7, column=1, pady=10)

root.mainloop()
    
```
